### **Data Cleaning**

```python
#duplicate check
print('Shape: ', data.shape, '\n')
data.duplicated().sum()
```

```python
#removing duplicate
data=data[~data.duplicated()]
print('Shape: ', data.shape, '\n')
print('Columns: ', data.columns.tolist(), '\n')
data.head()
```

```python
# Mapping old column names to new column names
column_mapping = {
    'dateCrawled': 'DateTime',
    'name': 'Car Name',
    'seller': 'Seller',
    'offerType': 'Offer Type',
    'price': 'Price',
    'abtest': 'AB Test',
    'vehicleType': 'Vehicle Type',
    'yearOfRegistration': 'Year of Registration',
    'gearbox': 'Gearbox',
    'powerPS': 'Power PS',
    'model': 'Model',
    'kilometer': 'Kilometer',
    'monthOfRegistration': 'Month of Registration',
    'fuelType': 'Fuel Type',
    'brand': 'Brand',
    'notRepairedDamage': 'Not Repaired Damage',
    'dateCreated': 'Date Created',
    'nrOfPictures': 'Pictures',
    'postalCode': 'Postal Code',
    'lastSeen': 'Last Seen'
}

# Rename columns using the mapping
data.rename(columns=column_mapping, inplace=True)

print('Shape: ', data.shape, '\n')
print('Columns: ', data.columns.tolist(), '\n')
```

```python
# Splitting date and time into two columns
data[['Date', 'Time']] = data['DateTime'].str.split(' ', 1, expand=True)

# Dropping the original datetime column if needed
data.drop(columns=['DateTime'], inplace=True)

print('Shape: ', data.shape, '\n')
print('Columns: ', data.columns.tolist(), '\n')
```

```python
data = data.apply(lambda x: x.str.title().str.replace('_', ' ') if x.dtype == "object" else x)
data.head()
```

```python
desired_columns = ['Date', 'Time', 'Car Name', 'Model', 'Vehicle Type', 'Year of Registration',
                   'Month of Registration', 'Seller', 'Offer Type', 'Price', 'AB Test', 'Gearbox',
                   'Power PS', 'Kilometer', 'Fuel Type', 'Brand', 'Not Repaired Damage',
                   'Date Created', 'Pictures', 'Postal Code', 'Last Seen']

# Reorder the columns
data = data.reindex(columns=desired_columns)

# Now you can save the DataFrame to a CSV file
data.to_csv(file, index=False)

print('Data Cleaning and File creation successful..!', file)
```
